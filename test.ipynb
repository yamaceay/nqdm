```python
arg1 = {k: v for k,v in zip(list("abcde"), list("fghij"))}
arg2 = [v**2 for v in range(10)]
arg3 = 4.0
```

```python
for data in nqdm(arg1):
    print(data)
```

```python
for data in nqdm(arg1, arg2, arg3):
    print(data)
```

```python
arg1_deep = np.arange(125).reshape(5, 5, 5)
arg2_deep = {str(i): {k+str(i): v+str(i) for k,v in zip(list("abcde"), list("fghij"))} for i in range(20)}
```

```python
for data in nqdm(arg1_deep, depth=0):
    print(data)
```

```python
for data in nqdm(arg1_deep, depth=1):
    print(data)
```

```python
for data in nqdm(arg1_deep, depth=2):
    print(data)
```

```python
for data in nqdm(arg2_deep, depth=0):
    print(data)
```

```python
for data in nqdm(arg2_deep, depth=1):
    print(data)
```

```python
arg1_bin = [2, 2, 2, 2]
```

```python
for data in nqdm(*arg1_bin, order="last"):
    print(data)
```

```python
for data in nqdm(*arg1_bin, order="first"):
    print(data)
```

```python
for data in nqdm(*arg1_bin, order=[0, 3, 1, 2]):
    print(data)
```

```python
for data in nqdm(*arg1_bin, order="last", enum=True):
    print(data)
```

```python
for data in nqdm(arg1_deep, arg2_deep, depth=[0, 1], order="last", enum=False):
    print(data)
```
